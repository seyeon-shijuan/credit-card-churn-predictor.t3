```python
df_drop_unknown=df[~(df["Education_Level"]=="Unknown")&~(df["Marital_Status"]=="Unknown")&~(df["Income_Category"]=="Unknown")]
```

```python
df_drop_unknown
```

```python
DF=df_drop_unknown.drop(["CLIENTNUM"],axis=1)
```

```python
DF=DF.drop(["Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_2"],axis=1)
```

```python
DF=DF.drop(["Naive_Bayes_Classifier_Attrition_Flag_Card_Category_Contacts_Count_12_mon_Dependent_count_Education_Level_Months_Inactive_12_mon_1"],axis=1)
```

```python
DF
```

```python
DF.info()
```

```python
DF["Attrition_Flag"]=DF["Attrition_Flag"].replace("Existing Customer", 0)
```

```python
DF["Attrition_Flag"]=DF["Attrition_Flag"].replace("Attrited Customer", 1)
```

```python
DF
```

```python
gender_dummies=pd.get_dummies(DF["Gender"],prefix="Gender")
```

```python
gender_dummies
```

```python
DF=pd.concat([DF,gender_dummies],axis=1)
```

```python
DF=DF.drop(["Gender"],axis=1)
```

```python
DF
```

```python
DF["Education_Level"].unique()
```

```python
replace_dict = {'Doctorate': 5, 'Post-Graduate': 4, 'Graduate': 3,'College':2,'High School':1, 'Uneducated':0}
DF["Education_Level"].replace(replace_dict, inplace=True)
```

```python
DF
```

```python
DF['Income_Category'].unique()
```

```python
replace_dict_2 = {'$120K +': 4, '$80K - $120K': 3, '$60K - $80K': 2,'$40K - $60K':1,'Less than $40K':0}
DF["Income_Category"].replace(replace_dict_2, inplace=True)
```

```python
DF
```

```python
Marital_dummies=pd.get_dummies(DF["Marital_Status"],prefix="Marital_Status")
```

```python
DF=pd.concat([DF,Marital_dummies],axis=1)
DF=DF.drop(["Marital_Status"],axis=1)
```

```python
DF
```

```python
DF['Card_Category'].unique()
```

```python
replace_dict_3 = {'Platinum': 3, 'Gold': 2, 'Silver': 1,'Blue':0}
DF["Card_Category"].replace(replace_dict_3, inplace=True)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

y=DF["Attrition_Flag"]
x=DF.drop(["Attrition_Flag"],axis=1)

X_train, X_test, y_train, y_test=train_test_split(x, y, test_size=0.2, random_state=11)

model=DecisionTreeClassifier(max_depth=3)

model.fit(X_train,y_train)

accuracy=model.score(X_test,y_test)

print(f'{accuracy=:.4f}')

import matplotlib.pyplot as plt
from sklearn import tree

plt.figure(figsize=(30,5))
tree.plot_tree(model,
               class_names=["Existing Customer","Attrited Customer"],
               feature_names=DF.columns[1:],
               filled=True,
               rounded=True)

plt.show()
```

```python
model=DecisionTreeClassifier(criterion="entropy",max_depth=5)

model.fit(X_train,y_train)

accuracy=model.score(X_test,y_test)

print(f'{accuracy=:.4f}')

import matplotlib.pyplot as plt
from sklearn import tree

plt.figure(figsize=(30,5))
tree.plot_tree(model,
               class_names=["Existing Customer","Attrited Customer"],
               feature_names=DF.columns[1:],
               filled=True,
               rounded=True)

plt.show()
```

```python
importances=model.feature_importances_

import numpy as np

indices=np.argsort(importances)
sorted_feature_names = [DF.columns[1:][i] for i in indices]
# Create plot
plt.figure()
plt.title("Feature Importance")
ticks = range(len(DF.columns[1:]))
plt.barh(ticks, importances[indices], align="center")
plt.yticks(ticks, sorted_feature_names)
plt.ylabel("Features")
plt.xlabel("Importance Score")
plt.tight_layout()
plt.show()
```
