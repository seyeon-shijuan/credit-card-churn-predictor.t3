## 범주형

### • Gender(성별)

```python
import pandas as pd

data = pd.read_csv("missing_value_processing_data_yc.csv")
F_Attrition_Flag = "Attrition_Flag"
data[F_Attrition_Flag] = data[F_Attrition_Flag].apply(lambda x: 0 if x == "Existing Customer" else 1)

F_Gender = "Gender"

data[[F_Attrition_Flag, F_Gender]].groupby(F_Gender).sum().sort_values("Attrition_Flag", ascending=False).plot.bar()
```

## 연속형

### • Customer_Age(고객 연령)

```python
import pandas as pd

data = pd.read_csv("missing_value_processing_data_yc.csv")
F_Attrition_Flag = "Attrition_Flag"
data[F_Attrition_Flag] = data[F_Attrition_Flag].apply(lambda x: 0 if x == "Existing Customer" else 1)

F_Customer_Age = "Customer_Age"

concat_data = pd.concat([data[F_Attrition_Flag], data[F_Customer_Age].apply(lambda x: x // 10)], axis="columns")
concat_data.groupby(F_Customer_Age).sum().sort_values("Customer_Age", ascending=True).plot.bar()
```

### • Dependent_count(부양가족수)

## 이상치

### Customer_Age(고객 연령)

```python
import pandas as pd

data = pd.read_csv("missing_value_processing_data_yc.csv")
F_Attrition_Flag = "Attrition_Flag"
data[F_Attrition_Flag] = data[F_Attrition_Flag].apply(lambda x: 0 if x == "Existing Customer" else 1)

F_Customer_Age = "Customer_Age"

Q1 = data[F_Customer_Age].quantile(q=0.25)
Q3 = data[F_Customer_Age].quantile(q=0.75)
IQR = Q3-Q1

min = Q1 - 1.5*IQR
max = Q3 + 1.5*IQR

print(f"Q1 = {Q1}, Q3 = {Q3}, IQR = {IQR}")
print(f"min = {min}, max = {max}")

data[F_Customer_Age].plot.box()
```
